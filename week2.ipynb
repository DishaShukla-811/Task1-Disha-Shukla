```
Q1 = df["TotalPrice"].quantile(0.25)
Q3 = df["TotalPrice"].quantile(0.75)

IQR = Q3 - Q1

outliers = df[(df["TotalPrice"] < (Q1 - 1.5*IQR)) |
              (df["TotalPrice"] > (Q3 + 1.5*IQR))]

outliers
```

```
df["Date"] = pd.to_datetime(df["Date"])

df.groupby(df["Date"].dt.month)["TotalPrice"].sum()
```
